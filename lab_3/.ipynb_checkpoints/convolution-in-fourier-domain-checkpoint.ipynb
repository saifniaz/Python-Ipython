# Convolution using Fourier Transform

```python
import numpy as np
import scipy as sp
from scipy import signal
import cv2
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('test.jpg', 0) # Load in grayscale

plt.figure(figsize=(5,5))
plt.imshow(img, 'gray')
```

**Convolving an image with a sobel filter** 

```python
sobel_x = np.arange(10000).reshape(100, 100)
start1= int(round(time.time()*1000))
img_x = sp.signal.convolve(img, sobel_x, mode='same')
end1 = int(round(time.time()*1000))
plt.figure(figsize=(5,5))
plt.imshow(img_x, 'gray')
temp = end1 - start1
print('Image Domain', temp, "ms")
```

**Performing the same convolution using `fftconvolve` method that uses Fourier Transform to perform convolutions**

```python
start2 = int(round(time.time()*1000))
img_x = sp.signal.fftconvolve(img, sobel_x, mode='same')
end2 = int(round(time.time()*1000))
plt.figure(figsize=(5,5))
plt.imshow(img_x, 'gray')
temp = end2 - start2
print('Frequency Domain', temp, "ms")
```

```python
tmp = np.zeros(img.shape)
tmp[0:100, 0:100] = sobel_x
start2 = int(round(time.time()*1000))
img_fft = np.fft.fft2(img)
tmp_fft = np.fft.fft2(tmp)
result = np.fft.ifft2( np.multiply(img_fft, tmp_fft) )
end2 = int(round(time.time()*1000))
plt.figure(figsize=(5,5))
plt.imshow(np.real(result), 'gray')

temp = end2 - start2
print('Frequency Domain', temp, "ms")
```

**Playing with large images**

```python
h, w = 10000, 10000
img = np.zeros([h, w])
img[:,w//2:] = 255

plt.figure(figsize=(5,5))
plt.imshow(img, 'gray')
```

```python
img_x = sp.signal.convolve(img, sobel_x, mode='same')

plt.figure(figsize=(5,5))
plt.imshow(img_x, 'gray')
```
